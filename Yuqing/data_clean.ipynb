```python
# extract ebill_enrolled_status_13_march and ebill_enrolled_status_18_march
# and compare the difference between the two columns, count the difference
# and calculate the percentage of the difference they are string type

data_test['ebill_enrolled_status_13_march'] = data_test['ebill_enrolled_status_13_march'].astype(str)
data_test['ebill_enrolled_status_18_march'] = data_test['ebill_enrolled_status_18_march'].astype(str)

data_test['diff'] = data_test['ebill_enrolled_status_13_march'] != data_test['ebill_enrolled_status_18_march']

data_test['diff'] = data_test['diff'].astype(int)

# print(data_test['diff'].sum() / len(data_test))

# extract the rows that have difference between ebill_enrolled_status_13_march and ebill_enrolled_status_18_march
# and print the difference

data_test_diff = data_test[data_test['diff'] == 1]

print(data_test_diff[['ebill_enrolled_status_13_march', 'ebill_enrolled_status_18_march']])


```

```python
# 'timestamp_call_key’: Date and time format when the customer called Synchrony. YYMMDDHHMMSS you can ignore the DAL/PHX and anything after.
#  only keep the first 6 characters of the timestamp_call_key
# data_test['timestamp_call_key'] = data_test['timestamp_call_key'].str[:6]
# # assign the retailer code to number each retailer code is a upper case letter, so we can convert it to a number
# data_test['retailer_code'] = data_test['retailer_code'].apply(lambda x: ord(x) - 64)

import matplotlib.pyplot as plt
import seaborn as sns





data_test['ebill_enrolled_status_13_march'], legend_ebill_enrolled_status_13_march       = pd.factorize(data_test['ebill_enrolled_status_13_march'])
data_test['account_status_13_march']        ,  legend_account_status_13_march      = pd.factorize(data_test['account_status_13_march'])
data_test['account_status_18_march']        ,    legend_account_status_18_march      = pd.factorize(data_test['account_status_18_march'])
data_test['delinquency_history_18_march']   ,  legend_delinquency_history_18_march  = pd.factorize(data_test['delinquency_history_18_march'])
data_test['ebill_enrolled_status_18_march'] ,  legend_ebill_enrolled_status_18_march   = pd.factorize(data_test['ebill_enrolled_status_18_march'])
data_test['delinquency_history_13_march']   ,  legend_delinquency_history_13_march      = pd.factorize(data_test['delinquency_history_13_march'])




conversion_legends = {
    'ebill_enrolled_status_13_march': legend_ebill_enrolled_status_13_march,
    'account_status_13_march':          legend_account_status_13_march,
    'account_status_18_march':            legend_account_status_18_march,
    'delinquency_history_18_march':     legend_delinquency_history_18_march,
    'ebill_enrolled_status_18_march':   legend_ebill_enrolled_status_18_march,
    'delinquency_history_13_march':     legend_delinquency_history_13_march,
}



print(data_test[['ebill_enrolled_status_13_march']].head())
for label, legend in conversion_legends.items():
    print(f"Conversion legend for {label}:")
    for i, value in enumerate(legend):
        print(f"{value} => {i}")
    print("\n")

sns.set_style("whitegrid")

fig, ax = plt.subplots(3, 2, figsize=(14, 12))
ax = ax.ravel()

variables = [
    'ebill_enrolled_status_13_march', 
    'account_status_13_march', 
    'account_status_18_march', 
    'delinquency_history_18_march', 
    'ebill_enrolled_status_18_march', 
    'delinquency_history_13_march'
]



for i, var in enumerate(variables):
    sns.countplot(data=data_test, x=var, ax=ax[i], order=data_test[var].value_counts().index)
    ax[i].set_title(var.replace('_', ' ').title())
    ax[i].set_xlabel(f'{var}')
    ax[i].set_ylabel('Count')
    for label in ax[i].get_xticklabels():
        label.set_rotation(45)



plt.tight_layout()
plt.show()


fig, ax = plt.subplots(3, 2, figsize=(14, 12))
ax = ax.ravel()

for i, var in enumerate(variables):
    counts = data_test[var].value_counts().reset_index()
    counts.columns = [var, 'counts'] 
    sns.scatterplot(ax=ax[i], data=counts, x=var, y='counts')
    ax[i].set_title(f'Scatter Plot of {var}')
    ax[i].set_xlabel('Encoded Label')
    ax[i].set_ylabel('Counts')
    ax[i].grid(True)

    
plt.show()

```
