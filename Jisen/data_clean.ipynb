```python
# import data from the csv file  F:\university\UIUC\24SPRING\datathon\file_3_Mar18_Output_1.csv 
import pandas as pd
import numpy as np

data = pd.read_csv('file_4_Mar18_Output_1.csv')
```

```python
data_test = data
```

```python
data_test["retailer_code"].value_counts()
data_test['retailer_code'], legend_retailer_code = pd.factorize(data_test['retailer_code'])
data_test['reason'], legend_reason = pd.factorize(data_test['reason'])
data_test['resolved'], legend_resolved = pd.factorize(data_test['resolved'])
# Zay
data_test['ebill_enrolled_status_13_march'], legend_ebill_enrolled_status_13_march = pd.factorize(
    data_test['ebill_enrolled_status_13_march'])
data_test['account_status_13_march'],  legend_account_status_13_march = pd.factorize(
    data_test['account_status_13_march'])
data_test['account_status_18_march'],    legend_account_status_18_march = pd.factorize(
    data_test['account_status_18_march'])
data_test['ebill_enrolled_status_18_march'],  legend_ebill_enrolled_status_18_march = pd.factorize(
    data_test['ebill_enrolled_status_18_march'])

conversion_legends = {
    'retailer_code': legend_retailer_code,
    'reason': legend_reason,
    'resolved': legend_resolved,
    # Zay
    'ebill_enrolled_status_13_march': legend_ebill_enrolled_status_13_march,
    'account_status_13_march':legend_account_status_13_march,
    'account_status_18_march':legend_account_status_18_march,
    'ebill_enrolled_status_18_march':legend_ebill_enrolled_status_18_march,
}

for label, legend in conversion_legends.items():
   print(f"Conversion legend for {label}:")
   for i, value in enumerate(legend):
       print(f"{value} => {i}")
   print("\n")
```

```python
def remove_(s):
    return int(s.strip("[]"))

data_test['delinquency_history_13_march'] = data_test['delinquency_history_13_march'].astype(str)
data_test['delinquency_history_18_march'] = data_test['delinquency_history_18_march'].astype(str)

data_test['delinquency_history_13_march'] = data_test['delinquency_history_13_march'].apply(
    remove_)
data_test['delinquency_history_18_march'] = data_test['delinquency_history_18_march'].apply(
    remove_)
```
